```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import r2_score
from tqdm import tqdm
%matplotlib inline
```

```python
df = pd.read_csv('detectedfooditems_modified.csv')
```

```python
df
```

```python
df.info()
```

```python
unique_entries_count = df['Name'].nunique()
unique_entries_count
```

```python
conditions = (df['Calories'] <= 600) & \
             (df['FatContent'] <= 200) & \
             (df['SaturatedFatContent'] <= 45) & \
             (df['CholesterolContent'] <= 100) & \
             (df['SodiumContent'] <= 900) & \
             (df['CarbohydrateContent'] <= 60) & \
             (df['FiberContent'] <= 13) & \
             (df['SugarContent'] <= 12) & \
             (df['ProteinContent'] <= 40)

# Apply conditions to filter out entries
filtered_df = df[conditions]

filtered_df
```

```python
# Define conditions
conditions = (df['Calories'] <= 600.0) & \
             (df['FatContent'] <= 200.0) & \
             (df['SaturatedFatContent'] <= 45.0) & \
             (df['CholesterolContent'] <= 100.0) & \
             (df['SodiumContent'] <= 900.0) & \
             (df['CarbohydrateContent'] <= 60.0) & \
             (df['FiberContent'] <= 13.0) & \
             (df['SugarContent'] <= 12.0) & \
             (df['ProteinContent'] <= 40.0)

# Create a new column and assign labels based on conditions
df['recommended'] = [0 if condition else 1 for condition in conditions]
```

```python
df
```

```python
df.to_csv("recommended.csv", index=False)
```

```python
df['recommended'] = [1 if condition else 0 for condition in conditions]

count_recommended = df['recommended'].value_counts().get(1, 0)
count_not_recommended = df['recommended'].value_counts().get(0, 0)

print("Recommended foods:", count_recommended)
print("Not recommended foods:", count_not_recommended)
```

```python
del df['foodnumber']
```

```python
pre_final = df.iloc[:,1:]
pre_final
```

```python
X = pre_final.drop(['recommended'], axis='columns')
```

```python
y = pre_final['recommended']
```

```python
#from sklearn.preprocessing import StandardScaler
#scaler=StandardScaler()
#X_data=scaler.fit_transform(X.to_numpy())
#X_data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=101)
```

```python
from sklearn import svm
clf = svm.SVC()
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(metrics.classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV 
parameters = {'C': [0.1,1,10,100,1000], 'gamma': [1,0.1,0.01,0.001,0.0001]}
```

```python
svm_clf = GridSearchCV(svm.SVC(),parameters,refit=True,verbose=2)
svm_clf.fit(X_train, y_train)
```

```python
svm_clf.best_params_
```

```python
means = svm_clf.cv_results_['mean_test_score']
stds = svm_clf.cv_results_['std_test_score']
for mean, std, params in zip(means, stds, svm_clf.cv_results_['params']):
    print("%0.3f (+/-%0.03f) for %r" % (mean, std * 2, params))
```

```python
grid_predictions = svm_clf.predict(X_test)
```

```python
print(metrics.classification_report(y_test, grid_predictions))
```

```python
metrics.accuracy_score(y_test, grid_predictions)
```

```python
print(confusion_matrix(y_test,grid_predictions))
```
